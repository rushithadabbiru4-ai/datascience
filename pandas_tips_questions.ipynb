#### Display message low if total_bill is less than 10, medium if less than 20 otherwise high

```python
def get_message(v):
    if v['total_bill'] < 10:
        return 'Low'
    elif v['total_bill'] < 20:
        return 'Medium'
    else:
        return 'High'

df['message'] = df.apply(get_message,axis=1)
df
```

#### Display top 3 transactions based on bill amount for each day

```python
day_group = df.groupby('day')
# First element is group and second one is row in that group
for g,trans in day_group:
    print(g)
    print(trans.sort_values(by='total_bill', ascending=False).head(3))
```

#### Display ratio of each day's total_bill to the total_bill for all days

```python
total = df['total_bill'].sum()
df.groupby('day')['total_bill'].sum()/total
```

#### Display max and min total_bill for each day

#### Display day on the left and sex at the top and display total bill amount at the intersection

```python
df.pivot_table(index='day',columns='sex',aggfunc=sum,values=['total_bill'])
```

#### Display total number of customers on each day for each time

```python
df.pivot_table(index='day',columns='time',aggfunc=sum,values=['size'])
```

#### Display day on left, sum and average of bill amount and tip amount for each sex

```python
df.pivot_table(index='day',columns='sex',aggfunc={'total_bill':['sum','mean'],'tip':['sum','mean']})
```
